```python
import rdflib
import json
```

```python
with open('cleaned_tweets.json','r') as jsonFile:
    data = json.load(jsonFile)

g = rdflib.Graph()

for i in range(len(data)):
    txt = data[i]
    try:
        s = rdflib.URIRef('http://localhost:8888/edit/Desktop/comp8210/assignment/ass2/cleaned_tweets.json' + txt['id'])
        p = rdflib.URIRef('http://localhost:8888/edit/Desktop/comp8210/assignment/ass2/cleaned_tweets.json' + txt['twitter_entities'])
        o = rdflib.URIRef('http://localhost:8888/edit/Desktop/comp8210/assignment/ass2/cleaned_tweets.json' + txt['actor'])
        g.add((s,p,o))
    except Exception as e:
        pass
g.serialize('data.rdf')

jsonFile.close()
```
